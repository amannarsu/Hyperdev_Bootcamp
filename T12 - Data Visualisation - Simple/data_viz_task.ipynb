```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False) 
```

### 1. Boxplot

```python
# Boxplot for revs per mile
#car_data.columns
plt.figure()
specific_Column = car_data[['Manufacturer', 'Rev.per.mile']]
filtered_Manufacturer = ['Audi','Hyundai','Suzuki','Toyota']
by_Manufacturer =specific_Column[specific_Column['Manufacturer'].isin(filtered_Manufacturer)]
plt.title("Revs per mile based on Manufacturer")
#sns.set_theme(style="ticks", palette="deep")
sns.boxplot(x="Manufacturer", y="Rev.per.mile", data= by_Manufacturer)
plt.show()
#print (f"Actual data for comparison:\n {by_Manufacturer}")
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Suziki has the highest revs per mile


### 2. Histogram

```python
# Histogram of MPG
#car_data.columns
plt.figure()
spg_Column2 = car_data[['MPG.city', 'MPG.highway']]
plt.subplot(1,2,1)
plt.title("City Mileage")
plt.xlabel("Mileage")
plt.ylabel("Gallon")
#plt.hist(car_data['MPG.city'], color='red')
sns.histplot(data=car_data, x="MPG.city", kde=True)
plt.subplot(1,2,2)
plt.title("Highway Mileage")
plt.xlabel("Mileage")
plt.ylabel("Gallon")
sns.histplot(data=car_data, x="MPG.highway", kde=True)
#plt.hist(car_data['MPG.highway'], color='green')
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
Highway is more fule efficient

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
car_data_sorted=car_data.sort_values(['Turn.circle'], ascending=[True]) #sorted(car_data['Max.Price'])
#turn_circle= car_data['Turn.circle']
#wheel_base= car_data['Wheelbase']
plt.figure()
plt.title("Wheelbase vs turning circle")
plt.xlabel("Wheelbase")
plt.ylabel("Turn.circle")
#plt.plot(wheel_base, turn_circle, c='g')
sns.set_theme(style="darkgrid", palette="pastel")
sns.lineplot(x='Wheelbase', y='Turn.circle', data=car_data)
plt.show()
#print(car_data_sorted)
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
The turning circle increases as the wheelbase gets larger

### 4. Barplot

```python
# Barplot for mean horsepower
car_data.columns
group_by_type = car_data.groupby(by=['Type'])
car_data_avg = round(group_by_type.mean(numeric_only=True), 0) 
#car_data_avg.sample(5)
sns.set_theme(style="whitegrid", palette="deep")
sns.barplot(car_data_avg, x="Type", y="Horsepower")
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
Yes, The graph above proves that larger car requires more horsepower
