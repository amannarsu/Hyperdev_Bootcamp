```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Import the dataset
df = pd.read_csv("insurance.csv")
```

```python
#BASIC CHECKS#

#check no of rows and columns
print(f"\nSize of rows/columns: {df.shape}\n")

#check details of the dataset
print(df.info(),"\n")

#checking for null values
print(f"\nChecking for if there are null values: \n{df.isnull().sum()}")

# Check the 1st 5 values in the dataset
print(f"\nChecking the first 5 Rows and all columns of data: \n{df.head()}\n")

df.describe()
```

```python
# Plot data in a scatter plot to view the data before liner regression
plt.figure()
sns.scatterplot(data=df, x='age', y='charges')
plt.show()
plt.close()
```

```python
X = df['age'].values.reshape(-1, 1) # create a 2D array
y = df['charges'].values.reshape(-1, 1)

# Fit and predict
# Create an empty linear regression model
simple_model = LinearRegression()

# Create the model using fit(x,y)
simple_model.fit(X, y)

# Use model to make a prediction
y_pred = simple_model.predict(X)

# Plot model and data
plt.figure()
sns.scatterplot(data=df, x='age', y='charges')
plt.plot(X, y_pred)
plt.show()
plt.close()
```

You can see a clear line for the Best fit showing that the charges for the insurance goes up based on the age in general

Lets check if the Charges are affected by BMI as well for the insurance

```python
# Plot data in a scatter plot to view the data before liner regression
plt.figure()
sns.scatterplot(data=df, x='bmi', y='charges')
plt.show()
plt.close()
```

```python
X2 = df['bmi'].values.reshape(-1, 1) # create a 2D array
y2 = df['charges'].values.reshape(-1, 1)

# Fit and predict
# Create an empty linear regression model
simple_model = LinearRegression()

# Create the model using fit(x,y)
simple_model.fit(X2, y2)

# Use model to make a prediction
y2_pred = simple_model.predict(X2)

# Plot model and data
plt.figure()
sns.scatterplot(data=df, x='bmi', y='charges')
plt.plot(X2, y2_pred)
plt.show()
plt.close()
```

Based on the best fit line in the scatterplot above, the cost of insurance is higher, the higher the BMI is for the customer
